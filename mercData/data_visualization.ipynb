```python
plt.figure(figsize=(10, 6))
plt.hist(prcmpg, bins=20, color='skyblue', edgecolor='black', alpha=0.7)  # Histogram çizimi, bins sayısını ve renkleri ayarlayabilirsiniz
plt.grid(True, linestyle='--', alpha=0.5)  # Izgara ekleyin
plt.title('Fiyat ve mil başına yakıt tüketimi arasındaki ilişki')
plt.xlabel('Fiyat')
plt.ylabel('Mil başına yakıt tüketimi')
plt.tight_layout()  # Grafik öğelerinin düzenini ayarlayın
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.savefig('Fiyat ve mil başına yakıt tüketimi arasındaki ilişki')
```
